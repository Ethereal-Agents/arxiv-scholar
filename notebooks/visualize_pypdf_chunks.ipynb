# PyPDF (Sliding Window) Chunk Visualization
This notebook allows you to load a single PDF from the local dataset and visualize exactly how the `SlidingWindowChunker` chops it up, so you can compare it visually to Docling.

```python
import os
import sys
import fitz  # PyMuPDF

# Add src to pythonpath so we can import the chunker
sys.path.insert(0, os.path.abspath('./src'))

from arxiv_scholar.chunking.sliding_window import SlidingWindowChunker
from arxiv_scholar.schema import Document
```

```python
def extract_text_pypdf(pdf_path):
    """Extracts raw text from a PDF using PyMuPDF (fitz), matching our local ingestion script."""
    text = ""
    try:
        with fitz.open(pdf_path) as doc:
            for page in doc:
                text += page.get_text() + "\n"
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
    return text

# Select a PDF from our dataset
pdf_dir = "/Users/tri/Projects/arxiv-scholar/nlp_ml_gcs_pdfs"
pdf_files = [f for f in os.listdir(pdf_dir) if f.endswith('.pdf')]
target_pdf = os.path.join(pdf_dir, pdf_files[0])  # Just grab the first one

print(f"Processing: {target_pdf}")
raw_text = extract_text_pypdf(target_pdf)
print(f"Total raw characters extracted: {len(raw_text)}")
```

```python
# Instantiate the Sliding Window Chunker (using the same settings as our pipeline)
chunker = SlidingWindowChunker(chunk_size=2000, chunk_overlap=200)

# Mock a Document object for the chunker
doc = Document(id="mock_id", source=target_pdf, content=raw_text)

chunks = list(chunker.chunk(doc))
print(f"Generated {len(chunks)} chunks!")
```

```python
def visualize_chunk(chunk, index):
    print("=" * 80)
    print(f"CHUNK {index} | Length: {len(chunk.content)} chars")
    print("=" * 80)
    print(chunk.content.strip())
    print("\n")

# Visualize the first 5 chunks
for i, chunk in enumerate(chunks[:5]):
    visualize_chunk(chunk, i)
```
